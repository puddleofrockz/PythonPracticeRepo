```python
import pandas as pd
import sqlite3
import os
os.getcwd()

db = sqlite3.connect("Desktop/Data/TripsDemo.db")

df1 = pd.read_csv("Desktop/Data/Trips.csv", header= 'infer')

df1.to_sql("Trips", db, if_exists= "replace")

def run_query (query):
    return pd.read_sql_query(query,db)
```
